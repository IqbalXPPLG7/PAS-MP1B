**Import DATA CSV**

```python
import pandas as pd
as_ = pd.read_csv('/content/drive/MyDrive/tugas/TShirt_size.csv')
as_.head()
```

```python
as_.rename(columns={
    'Height (in cms)': 'Height',
    'Weight (in kgs)': 'Weight',
    'T Shirt Size': 'Shirt'
}, inplace=True)

print(as_.head())
```

**Memvisualisasi DATASET**

```python
import matplotlib.pyplot as ptt
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for T_Shirt_Size, d in as_.groupby('Shirt'):
  ax.scatter(d['Height'], d['Weight'], label=T_Shirt_Size)

plt.legend(loc='upper left')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Scatter Plot Height vs Weight')
plt.grid(True)
plt.show()
```

**Preporcesing Data**

```python
import numpy as np

x_train = np.array(as_[['Height', 'Weight']])
y_train = np.array(as_['Shirt'])

print(f'x_train = {x_train}\n')
print(f'y_train = {y_train}')
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
y_train = lb.fit_transform(y_train)
print(f'y_train:\n = {y_train}')
```

**Memprediksi 3 tetangga**

```python
from sklearn.neighbors import KNeighborsClassifier

k = 3
model = KNeighborsClassifier(n_neighbors=k)
model.fit(x_train, y_train)
```

**Prediksi size BAJU**

```python
weight = 159
height = 60
X_new = np.array([height, weight]).reshape(1, -1)  # Reshape to 1 row and let NumPy infer the number of columns
X_new
```

```python
y_new = model.predict(X_new)
y_new
```

```python
lb.inverse_transform(y_new)
```

```python
X_test = np.array([[168, 63], [175, 66], [162, 52], [155, 51]])
y_test = lb.transform(np.array(['M', 'M', 'L', 'L'])).flatten()

print(f'X_test:\n{X_test}\n')
print(f'y_test:\n{y_test}')
```

```python
y_pred = model.predict(X_test)
y_pred
```

## **ACCURACY**

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score (y_test, y_pred)
print(f'Accuracy: {acc}')
```

## **PRECISSION**

```python
from sklearn.metrics import precision_score
prec = precision_score (y_test, y_pred)
print(f'Precission: {prec}')
```

## **RECALL**

```python
from sklearn.metrics import recall_score
rec = recall_score(y_test, y_pred)
print(f'Recall: {rec}')
```

## **F1 SCORE**

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
print(f'F1-score: {f1}')
```

## **CLASSIFICATION REPORT**

```python
from sklearn.metrics import classification_report
cls_report = classification_report(y_test, y_pred)
print(f'Classification Report:\n{cls_report}')
```

## **MCC**

```python
from sklearn.metrics import matthews_corrcoef

mcc = matthews_corrcoef(y_test, y_pred)

print(f'MCC: {mcc}')
```
